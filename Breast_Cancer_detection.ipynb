```python
## to shuffle data  to get effective result and accuacy
```

```python
df=pd.DataFrame(data['data'],columns=data['feature_names'])
df_target=pd.DataFrame(data['target'],columns=['Result'])
df_cct=pd.concat([df,df_target],axis=1)
shuffled_df=df_cct.sample(frac=1)
```

```python

shuffled_df
```

```python
data['data']=shuffled_df.values[:,:30]
```

```python
data['target']=shuffled_df.values[:,30:]
```

# Training validation and testing the machines



 we have 569 rows 



so training 500 rows

validation 35 rows

testing 34 rows

```python
tumor_data=data['data']
tumor_result=data['target']
```

```python
x_train=tumor_data[:500]
y_train=tumor_result[:500]


x_valid=tumor_data[500:535]
y_valid=tumor_result[500:535]


x_test=tumor_data[535:]
y_test=tumor_result[535:]
```

### Creating 1sthideen layer with 15 inputs

```python
model=Sequential()
model.add(Dense(64 , activation='relu'  ,input_dim=30))
model.add(Dense(104,activation='relu'))
model.add(Dense(14,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
## batch size is something wihch will be used to send data odf size and update the output
## by default the batch size is 32
## therefore  500/32=15.99
## therefore by defaaut the number of the batches is 16 for training 500 data
```

```python
model.fit(x_train,y_train,batch_size=1,epochs=20,validation_data=(x_valid,y_valid))
```

```python
model.evaluate(x_test,y_test)
```

```python
for i in range(0,34):
  sample= x_test[i]
  sample=np.reshape(sample,(1,30))

  if model.predict(sample)[0][0]>0.5:
    print('-benign')
  else :
    print('-malignant')
  if y_test[i]==0:
    print('*Malignant')
  else:
    print('*Benign')
  print('________________')
```
